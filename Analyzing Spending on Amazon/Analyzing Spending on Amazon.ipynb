```python
import pandas as pd

df = pd.read_csv('amazon-orders.csv')
df.head()
```

```python
df.shape
```

```python
df = df.fillna(0)
df.head()
```

```python
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
# calculating the total amount of money spent on Amazon
df['Total Charged'].sum()
```

```python
# calculating the average total spend on Amazon
df['Total Charged'].mean()
```

My average spend on Amazon is about $30!

```python
df['Total Charged'].median()
```

My median spend is $15.95, which is lower than my average. Seems like I've made a few pricey orders. Let's find out the most expensive and cheapest purchase I've made. 

```python
df['Total Charged'].max()
```

```python
df['Total Charged'].min()
```

```python
# calculating the amount of salex tax that I've paid
df['Tax Charged'] = df['Tax Charged'].str.replace('$', '').astype(float)
df.head()
```

```python
df['Tax Charged'].sum()
```

About $52.60 of the total money I spent on Amazon went towards taxes. 

```python
# The overall effective sales tax rate of about 2.9%.
df["Tax Charged"].sum() / df["Total Charged"].sum()
```

```python
# Analzing Amazon spending overtime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
%matplotlib inline
```

```python
df.plot.bar(x='Order Date', y='Total Charged', rot=90, figsize=(20,10))
```

```python
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
daily_orders.plot.bar(figsize=(20,10))
```
